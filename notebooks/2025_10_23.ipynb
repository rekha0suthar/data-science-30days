# Task
Analyze the numerical columns of the dataframe by calculating their covariance and correlation matrices, and visualize the relationships between them using scatterplots.

## Calculate covariance

### Subtask:
Calculate the covariance matrix for the numerical columns.


**Reasoning**:
Select the numerical columns and calculate the covariance matrix.



```python
numerical_cols = df.select_dtypes(include=['number'])
covariance_matrix = numerical_cols.cov()
display(covariance_matrix)
```

## Calculate correlation

### Subtask:
Calculate the correlation matrix for the numerical columns.


**Reasoning**:
Calculate the correlation matrix for the numerical columns and display it.



```python
correlation_matrix = numerical_cols.corr()
display(correlation_matrix)
```

## Visualize scatterplot

### Subtask:
Draw scatterplots to visualize the relationships between the numerical columns.


**Reasoning**:
Generate a pairplot to visualize the relationships between all pairs of numerical columns in the dataframe.



```python
sns.pairplot(numerical_cols)
```

## Summary:

### Correlation Matrix:
* There is a strong positive correlation between petal_length and petal_width (0.96). This suggests that as the length of the petal increases, the width also tends to increase.
* There is also a strong positive correlation between sepal_length and petal_length (0.87), and sepal_length and petal_width (0.82).
sepal_width shows weak or negative correlations with the other features.
### Scatterplots (Pairplot):
* The scatterplots visually confirm the relationships seen in the correlation matrix.
* The plots involving petal_length and petal_width show a clear linear relationship, with distinct clusters corresponding to the different species.
* The plots involving sepal_length with petal_length and petal_width also show positive trends, with some separation by species.
* The scatterplots involving sepal_width show less clear relationships with the other features, consistent with the low correlation values.
* The pairplot also shows the distribution of each feature on the diagonal.

In summary, the petal dimensions (petal_length and petal_width) are highly correlated and are good indicators for distinguishing between iris species. Sepal length is also somewhat correlated with petal dimensions. Sepal width is less correlated with the other features.


```python
import matplotlib.pyplot as plt

# Create scatterplots for each numerical feature against the 'species'
numerical_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=col, y='species', hue='species')
    plt.title(f'Scatterplot of {col} vs Species')
    plt.xlabel(col)
    plt.ylabel('Species')
    plt.show()
```
